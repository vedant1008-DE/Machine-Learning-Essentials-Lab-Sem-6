Question 1 — Load Datasets

```python
from sklearn.datasets import load_digits, load_breast_cancer
import pandas as pd

# Load datasets
digits = load_digits()
cancer = load_breast_cancer()

# Convert to DataFrame
digits_df = pd.DataFrame(digits.data)
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Display shapes
print("Digits Shape:", digits.data.shape)
print("Cancer Shape:", cancer.data.shape)

# Feature names
print("Cancer Feature Names:", cancer.feature_names)

# Sample output
print("Digits Sample:", digits.data[:5])
print("Cancer Sample:", cancer.data[:5])
```

Question 2 — Data Splitting

```python
from sklearn.model_selection import train_test_split

# Digits split
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    digits.data, digits.target, test_size=0.3, random_state=42)

# Cancer split
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    cancer.data, cancer.target, test_size=0.3, random_state=42)
```

Question 3 — Logistic Regression (Binary)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model_lr = LogisticRegression(max_iter=5000)
model_lr.fit(X_train_c, y_train_c)

y_pred_lr = model_lr.predict(X_test_c)

print("Accuracy:", accuracy_score(y_test_c, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test_c, y_pred_lr))
```

Question 4 — F1-Score (Binary)

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision:", precision_score(y_test_c, y_pred_lr))
print("Recall:", recall_score(y_test_c, y_pred_lr))
print("F1 Score:", f1_score(y_test_c, y_pred_lr))
```
