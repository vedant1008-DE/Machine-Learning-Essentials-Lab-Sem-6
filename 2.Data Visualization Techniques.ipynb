```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0:'setosa',1:'versicolor',2:'virginica'})

df_iris.head()
```

```python
sns.scatterplot(x='sepal length (cm)',
                y='sepal width (cm)',
                hue='species',
                data=df_iris)

plt.title("Sepal Length vs Width")
plt.show()
```

```python
df_iris.hist(figsize=(8,6))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(df_iris.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.violinplot(x='species',
               y='petal length (cm)',
               data=df_iris)

plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
sns.countplot(x='species', data=df_iris)
plt.title("Species Distribution")
plt.show()
```
